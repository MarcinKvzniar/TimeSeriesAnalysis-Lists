# Simple exponential smoothing (SES) - Grid search for simple exponential smoothing

Import dataset

```python
import pandas as pd
import numpy as np

df = pd.read_csv('AlgeriaExport.csv')
df['Year'] = pd.RangeIndex(start=1960, stop=1960+len(df))
df.set_index('Year', inplace=True)
df.tail()
```

Plot of raw data

```python
import matplotlib.pyplot as plt

plt.title("Exports: Algeria")
plt.ylabel("% of GDP")
plt.xlabel("Year")
plt.xlim(1955, 2020)
df['% of GDP'].plot()
```

Split the data into the training and testing parts

```python
import math
train_len = math.floor(len(df)*0.8)
train_data=df.iloc[:train_len]
test_data=df.iloc[train_len:]
```

Function to perform simple exponential smoothing (SES)

```python
def ses_forecast(df, alpha):
    forecasts = [df.iloc[0, 0]]  
    for i in range(1, len(df)):
        forecast = alpha * df.iloc[i, 0] + (1 - alpha) * forecasts[-1]
        forecasts.append(forecast)
    return forecasts
```

Function to calculate SSE

```python
def calculate_sse(actual, forecast):
    return np.sum((actual - forecast)**2)
```

Calculate SSE for each alpha

```python
alphas = np.linspace(0, 1, 100)
sse_values = []
for alpha in alphas:
    forecasts = ses_forecast(train_data, alpha)
    sse = calculate_sse(train_data.values.flatten()[1:], forecasts[:-1])  
    sse_values.append(sse)
```

Find the alpha with the smallest SSE and perform SES forecasting for it

```python
best_alpha = alphas[np.argmin(sse_values)]
best_forecasts = ses_forecast(df, best_alpha)
```

Plotting

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index, df.values, label='Original Data')
plt.plot(train_data.index, train_data.values, label='Training Data', linestyle='--')
plt.plot(test_data.index, test_data.values, label='Testing Data', linestyle='--')
plt.plot(df.index, best_forecasts, label='Forecast', linestyle='--')
plt.xlabel('Year')
plt.ylabel('% of GDP')
plt.title('Simple Exponential Smoothing Forecasting')
plt.legend()
plt.grid(True)
plt.show()
```
